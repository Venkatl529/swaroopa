```python
import tensorflow as tf
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
```

```python
feature_a = np.linspace(-1, 1, 101)
target_b = (5- (4 * feature_a) + (pow(feature_a,2)) + (3 * pow(feature_a,3)) - (2 * pow(feature_a,4)) + (pow(feature_a,5))+ np.random.randn(*feature_a.shape) * 0.25 )
```

```python
plt.figure(figsize = (6,6))
plt.title("Original data")
plt.scatter(feature_a,target_b)
```

```python
plt.figure(figsize=(6,6))

plt.title("Original data")
plt.plot(feature_a,target_b)
```

```python
class MyModel:
    def __call__(self, a):
        return self.w[0]
        + tf.multiply(a ,self.w[1])
        + tf.multiply(tf.pow(a,2),self.w[2])
        + tf.multiply(tf.pow(a,3),self.w[3])
        + tf.multiply(tf.pow(a,4),self.w[4])
        + tf.multiply(tf.pow(a,5),self.w[5]) 
    
    def __init__(self):
        self.w = tf.Variable([5.0,-4.0,1.0,3.0,-2.0,1.0])

    
```

# Loss function

```python
def loss(b, pred):
    return tf.reduce_mean(tf.square(b - pred))
```

# Hyper Parameters

```python
learning_rate = 0.01
epochs = 1000
```

# One step in the learning

```python
def one_step(model, a, b, learning_rate):
    with tf.GradientTape() as t:
        current_loss = loss(b, model(a))
    delta = t.gradient(current_loss, model.w)
    model.w.assign_sub(learning_rate * delta)
```

# Training to learn w

```python
model = MyModel()

Weights = []

for epoch_count in range(epochs):
    
    Weights.append(model.w.numpy()) 
    
    real_loss = loss(target_b, model(feature_a))
    
    one_step(model, feature_a, target_b, learning_rate)
    
    print(f"Epoch count {epoch_count}: Loss value: {real_loss.numpy()}")
```

```python
learned_w  =  Weights[-1]
print(learned_w)
```

```python
plt.figure(3)

plt.title("Regression line over data")

plt.xlabel("a values")
plt.ylabel("b values")

# original data
plt.scatter(feature_a, target_b,label="data")

#  predicted y values
predictions = 0
for i in range(learned_w.shape[0]):
    predictions += (feature_a ** i) * learned_w[i]

#  regression line 
plt.plot(feature_a, predictions, 'r', label="predictions")
plt.legend()
plt.show()
```
